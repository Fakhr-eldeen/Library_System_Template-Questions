```python
import mysql.connector
```

```python
mydb = mysql.connector.connect(
  host="localhost",
  user="root",
  password="admin",
  database="library",
  port = "3306"
)
print(mydb)
```

# find the number of availalbe copies of Dracula

```python
mycursor = mydb.cursor()
mycursor.execute("""
select count(distinct(b.BookID)) as books from books as b
right join  loans as l on l.BookID = b.BookID 
where b.Title = "Dracula"  and ReturnedDate = DueDate; 
""")
res = mycursor.fetchall()
res
```

# check total copies of the book

```python
mycursor2 = mydb.cursor()
mycursor2.execute("""
select count(distinct(b.BookID)) as books from books as b
cross join  loans as l on l.BookID = b.BookID 
where b.Title = "Dracula" ;
""")
res = mycursor2.fetchall()
res
```

# current total loans of the book

```python
mycursor3 = mydb.cursor()
mycursor3.execute("""
select count(l.BookID) from books as b
right outer join  loans as l on l.BookID = b.BookID 
where b.Title = "Dracula";
""")
res = mycursor3.fetchall()
res
```

# total available book

```python
mycursor4 = mydb.cursor()
mycursor4.execute("""
select count(distinct l.BookID) as Avialabel_books from books as b
left join  loans as l on l.BookID = b.BookID 
where b.Title = "Dracula";;
""")
res = mycursor4.fetchall()
res
```
