**What is the average Age ?**

```python
df["Age"].mean()
```

**What is the greatest Height? and name of the player with that height?**

```python
# greatest height
df["Height"].max()
```

```python
# name of the player
heighest_players =  df[df["Height"] == df["Height"].max()]
heighest_players["Name"]
```

**What is the smallest weight? and name of the player with that weight? and for which club plays?**

```python
df["Weight"].min()
```

```python
# name of the player
thinest_players = df[df["Weight"]== df["Weight"].min()]
thinest_players[["Name","Club"]]
```

**What is the position of Cristiano Ronaldo?**

```python
cr = df[df["Name"]=="Cristiano Ronaldo"]
cr["Position"]
```

**What was the average (mean) Value of all players per year(Joined column) ?**

```python
df.groupby("Joined")[['Value']].mean()
```

**get Number of players for each club**

```python
df.groupby("Club")[['Name']].count()
```

**How many unique Clubs are there?**

```python
df["Club"].nunique()
```

**What are the top 5 most common positions?**

```python
df.groupby("Position")[['Position']].count().sort_values(by='Position', ascending=False).head()
```

**What are the top 5 most common nationality?**

```python
df.groupby("Nationality")[['Nationality']].count().sort_values(by='Nationality', ascending=False).head()
```

**What are the names of players has Egyptian Nationality?**

```python
Egyptian_players = df[df["Nationality"]== "Egypt"]
Egyptian_players["Name"]
```

**How many players joined in 2018 with skill moves=2?** 

```python
 skilles = df[(df["Joined"]== 2018) & (df["Skill Moves"]==2)].count()
 skilles["Name"]
```

**How many players whom name starts with letter  'M'? (This is pretty tricky)**

```python
m = df[df['Name'].str.startswith('M')].count()
m["Name"]
```
